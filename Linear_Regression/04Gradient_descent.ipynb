```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
```

```python
# Applying OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color ='red')
```

```python
# Applying Gradient Descent assuming slope is constant m 78.35
# Let's assume that we are stating from intercept b=0
y_pred = ((78.35 * x) + 100).reshape(4)
```

```python
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color= 'red', label= 'OLS')
plt.plot(x, y_pred, color='blue', label='b=0')
plt.legend()
plt.show()
```

```python
m= 78.35
b=0
loss_slope = -2 *np.sum(y- m*x.ravel()-b) # the .ravel() method is used with NumPy arrays to flatten them into a 1D array. It provides a flattened view of the original array
loss_slope
```

```python
lr= 0.1
step_size = loss_slope *lr
step_size
```

```python
b= b- step_size
b
```

```python
y_pred1=((78.35* x) +b).reshape(4)
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color='red' , label='OLS')
plt.plot(x, y_pred1, color='blue', label='b={}'.format(b))
plt.plot(x, y_pred, color= 'green', label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = loss_slope *lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1=((78.35* x) +b).reshape(4)
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color='red' , label='OLS')
plt.plot(x, y_pred1, color='blue', label='b={}'.format(b))
plt.plot(x, y_pred, color= 'green', label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size= loss_slope*lr
step_size
```

```python
b= b - step_size
b
```

```python
y_pred1=((78.35* x) +b).reshape(4)
plt.scatter(x, y)
plt.plot(x, reg.predict(x), color='red' , label='OLS')
plt.plot(x, y_pred1, color='blue', label='b={}'.format(b))
plt.plot(x, y_pred, color= 'green', label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration in loop
b= -100
m = 78.35
lr = 0.01
epochs =100
for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*x.ravel() -b )
    b = b - (lr * loss_slope)
    y_pred = m*x+b
    plt.plot(x, y_pred)
plt.scatter(x, y)
```
