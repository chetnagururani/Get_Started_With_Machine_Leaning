# SGDClassifier

```python
from sklearn.datasets import load_diabetes
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
diabetes = load_diabetes()
# Convert to DataFrame
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
# Add the target column
df['diabetes_progression'] = diabetes.target
```

```python
X, y = load_diabetes(return_X_y= True)
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y,
                                                   test_size= 0.2 , random_state =2)
```

```python
df.head()
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
reg = SGDRegressor(max_iter=1000, learning_rate = 'constant', eta0=0.01)
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred= reg.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

loss: Type of loss function to use, like 'hinge' (for SVM) or 'log' (for logistic regression).

penalty: Type of regularization to apply, like 'l2' (Ridge) or 'l1' (Lasso) to prevent overfitting.

alpha: Controls the regularization strength; higher values mean stronger regularization.

l1_ratio: Used when penalty is 'elasticnet', controls the mix between l1 and l2 penalties.

fit_intercept: Whether to calculate the intercept (True) or set it to zero (False).

max_iter: Maximum number of iterations to run the solver.

tol: Tolerance for stopping criteria.

shuffle: Whether to shuffle the training data before each epoch.

verbose: Controls the level of verbosity during fitting (how much info gets printed out).

epsilon: Epsilon in the epsilon-insensitive loss functions; only for 'epsilon_insensitive' and 'squared_epsilon_insensitive'.
n_jobs: Number of CPU cores to use; -1 means using all available cores.

random_state: Seed for the random number generator (for reproducibility).

learning_rate: Strategy for updating the learning rate, 'optimal' is recommended.

eta0: Initial learning rate.

power_t: Exponent for inverse scaling learning rate.

early_stopping: Whether to use early stopping to terminate training when validation score is not improving.

validation_fraction: Fraction of training data to set aside as validation set.

n_iter_no_change: Number of iterations with no improvement to wait before stopping.

class_weight: Weights associated with classes, used for handling imbalanced datasets.

warm_start: Whether to reuse the solution of the previous call to fit as initialization for the next call.

average: Whether to average the coefficients of each iteration.
