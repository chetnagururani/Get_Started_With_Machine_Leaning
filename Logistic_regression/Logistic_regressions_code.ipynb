# LogisticRegression



```python
import pandas as pd
from sklearn.datasets import  load_iris
from sklearn.linear_model import LogisticRegression
X, y = load_iris(return_X_y= True)
```

```python
df = pd.DataFrame(X)
df.head()
```

```python
clf = LogisticRegression(max_iter=500,random_state=0)
```

```python
clf.fit(X, y)
```

```python
clf.predict(X[:2, :])
```

```python
clf.predict_proba(X[:2, :])
```

 model is a classification model, .score() calls accuracy_score(y_true, y_pred) internally.
 

```python
clf.score(X, y)
```

# GaussianNB


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
X, y =load_iris(return_X_y=True) 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

```python
gnb = GaussianNB()
```

```python
y_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
```

# LinearDiscriminantAnalysis
- LinearDiscriminantAnalysis (LDA) is a supervised dimensionality reduction technique and a classifier used in machine learning for classification tasks.

- It reduces the dimensionality while preserving class separability, making it useful when dealing with high-dimensional data.

  

```python
import numpy as np 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
y = np.array([1, 1, 1, 2, 2, 2])
```

```python
clf = LinearDiscriminantAnalysis()
```

```python
clf.fit(X, y)
```

```python
print(clf.predict([[-0.8, -1]]))
```
